```python
# SOAL 11.1  BAGIAN (a) - THOMAS ALGORITHM

print("SOAL 11.1 BAGIAN (a) - THOMAS ALGORITHM")
print()

# Koefisien
e = [0,    -0.4, -0.4]  # subdiagonal
f = [0.8,   0.8,  0.8]  # diagonal utama
g = [-0.4, -0.4,  0  ]  # superdiagonal
r = [41,    25,   105 ]  # ruas kanan

# FORWARD DECOMPOSITION
print("Langkah 1: Forward Decomposition")
print()

faktor1 = e[1] / f[0]
f[1] = f[1] - faktor1 * g[0]
r[1] = r[1] - faktor1 * r[0]
print(f"  k=2: faktor={faktor1:.4f}, f[1]={f[1]:.4f}, r[1]={r[1]:.4f}")

faktor2 = e[2] / f[1]
f[2] = f[2] - faktor2 * g[1]
r[2] = r[2] - faktor2 * r[1]
print(f"  k=3: faktor={faktor2:.4f}, f[2]={f[2]:.4f}, r[2]={r[2]:.4f}")

print()
print(f"  Matriks L: e={e}, f={[round(v,4) for v in f]}")
print(f"  Matriks U: g={g}")

# --- BACK SUBSTITUTION ---
print()
print("Langkah 2: Back Substitution")
print()

x3 = r[2] / f[2]
x2 = (r[1] - g[1] * x3) / f[1]
x1 = (r[0] - g[0] * x2) / f[0]

print(f"  x3 = {r[2]:.4f} / {f[2]:.4f} = {x3:.4f}")
print(f"  x2 = ({r[1]:.4f} - ({g[1]}) * {x3:.4f}) / {f[1]:.4f} = {x2:.4f}")
print(f"  x1 = ({r[0]:.4f} - ({g[0]}) * {x2:.4f}) / {f[0]:.4f} = {x1:.4f}")

print()
print(f"Solusi: x1={x1:.4f}, x2={x2:.4f}, x3={x3:.4f}")
```

```python
# SOAL 11.2 - INVERS MATRIKS MENGGUNAKAN LU DECOMPOSITION

print("SOAL 11.2 - INVERS MATRIKS MENGGUNAKAN LU DECOMPOSITION")
print()

e = [0,    -0.4, -0.4]
f = [0.8,   0.8,  0.8]
g = [-0.4, -0.4,  0.0]

# LANGKAH 1: LU DECOMPOSITION
print("Langkah 1: LU Decomposition")
print()

faktor1 = e[1] / f[0]
f[1] = f[1] - faktor1 * g[0]

faktor2 = e[2] / f[1]
f[2] = f[2] - faktor2 * g[1]

print(f"  faktor1 = {faktor1:.4f},  f[1] = {f[1]:.4f}")
print(f"  faktor2 = {faktor2:.4f},  f[2] = {f[2]:.4f}")
print()

# LANGKAH 2: UNIT VECTOR 1 = {1, 0, 0}
print("Langkah 2: Unit vector 1 = [1, 0, 0]")
print()

d1 = 1 / f[0]
d2 = (0 - e[1] * d1) / f[1]
d3 = (0 - e[2] * d2) / f[2]

x13 = d3
x12 = (d2 - g[1] * x13) / 1
x11 = (d1 - g[0] * x12) / 1

print(f"  d1={d1:.4f}, d2={d2:.4f}, d3={d3:.4f}")
print(f"  x1={x11:.4f}, x2={x12:.4f}, x3={x13:.4f}")
print()

# LANGKAH 3: UNIT VECTOR 2 = {0, 1, 0}
print("Langkah 3: Unit vector 2 = [0, 1, 0]")
print()

d1 = 0 / f[0]
d2 = (1 - e[1] * d1) / f[1]
d3 = (0 - e[2] * d2) / f[2]

x23 = d3
x22 = (d2 - g[1] * x23) / 1
x21 = (d1 - g[0] * x22) / 1

print(f"  d1={d1:.4f}, d2={d2:.4f}, d3={d3:.4f}")
print(f"  x1={x21:.4f}, x2={x22:.4f}, x3={x23:.4f}")
print()

# LANGKAH 4: UNIT VECTOR 3 = {0, 0, 1}
print("Langkah 4: Unit vector 3 = [0, 0, 1]")
print()

d1 = 0 / f[0]
d2 = (0 - e[1] * d1) / f[1]
d3 = (1 - e[2] * d2) / f[2]

x33 = d3
x32 = (d2 - g[1] * x33) / 1
x31 = (d1 - g[0] * x32) / 1

print(f"  d1={d1:.4f}, d2={d2:.4f}, d3={d3:.4f}")
print(f"  x1={x31:.4f}, x2={x32:.4f}, x3={x33:.4f}")
print()

# HASIL INVERS
print("Matriks Invers [A]^-1:")
print(f"  [{round(x11,4)},  {round(x21,4)},  {round(x31,4)}]")
print(f"  [{round(x12,4)},  {round(x22,4)},  {round(x32,4)}]")
print(f"  [{round(x13,4)},  {round(x23,4)},  {round(x33,4)}]")
```

```python
# SOAL 11.3 - THOMAS ALGORITHM SISTEM 4x4

print("SOAL 11.3 - THOMAS ALGORITHM SISTEM 4x4")
print()

e = [0,          -0.020875, -0.020875, -0.020875]
f = [2.01475,     2.01475,   2.01475,   2.01475 ]
g = [-0.020875,  -0.020875, -0.020875,  0       ]
r = [4.175,       0,         0,         2.0875  ]

n = 4

# FORWARD DECOMPOSITION
print("Langkah 1: Forward Decomposition")
print()

for k in range(1, n):
    faktor = e[k] / f[k-1]
    f[k] = f[k] - faktor * g[k-1]
    r[k] = r[k] - faktor * r[k-1]
    print(f"  k={k+1}: faktor={faktor:.6f}, f[{k}]={f[k]:.6f}, r[{k}]={r[k]:.6f}")

print()

# BACK SUBSTITUTION
print("Langkah 2: Back Substitution")
print()

T4 = r[3] / f[3]
T3 = (r[2] - g[2] * T4) / f[2]
T2 = (r[1] - g[1] * T3) / f[1]
T1 = (r[0] - g[0] * T2) / f[0]

print(f"  T4 = {r[3]:.6f} / {f[3]:.6f} = {T4:.4f}")
print(f"  T3 = ({r[2]:.6f} - ({g[2]}) * {T4:.4f}) / {f[2]:.6f} = {T3:.4f}")
print(f"  T2 = ({r[1]:.6f} - ({g[1]}) * {T3:.4f}) / {f[1]:.6f} = {T2:.4f}")
print(f"  T1 = ({r[0]:.6f} - ({g[0]}) * {T2:.4f}) / {f[0]:.6f} = {T1:.4f}")

print()
print(f"Solusi: T1={T1:.4f}, T2={T2:.4f}, T3={T3:.4f}, T4={T4:.4f}")
```

```python
# SOAL 11.4 - VERIFIKASI CHOLESKY DECOMPOSITION

print("SOAL 11.4 - VERIFIKASI CHOLESKY DECOMPOSITION")
print()

# Matriks A asli
A = [[6,   15,  55 ],
     [15,  55,  225],
     [55,  225, 979]]

# Matriks L dari Example 11.2
L = [[2.4495,  0,      0     ],
     [6.1237,  4.1833, 0     ],
     [22.454,  20.917, 6.1101]]

print("Matriks L:")
for row in L:
    print(f"  {[round(v, 4) for v in row]}")
print()

# Transpose L
LT = [[L[0][0], L[1][0], L[2][0]],
      [L[0][1], L[1][1], L[2][1]],
      [L[0][2], L[1][2], L[2][2]]]

print("Matriks L^T:")
for row in LT:
    print(f"  {[round(v, 4) for v in row]}")
print()

# Langkah 1: Hitung L x LT
print("Langkah 1: Hitung [L] x [L]^T")
print()

hasil = [[0.0]*3 for _ in range(3)]
for i in range(3):
    for j in range(3):
        hasil[i][j] = round(L[i][0]*LT[0][j] + L[i][1]*LT[1][j] + L[i][2]*LT[2][j], 4)

for row in hasil:
    print(f"  {row}")
print()

# Langkah 2: Bandingkan dengan A
print("Langkah 2: Bandingkan dengan matriks A asli")
print()
for row in A:
    print(f"  {row}")
print()

# Cek selisih
cocok = True
for i in range(3):
    for j in range(3):
        if abs(hasil[i][j] - A[i][j]) > 0.01:
            cocok = False

if cocok:
    print("Verifikasi BERHASIL: [L][L]^T = [A] terbukti benar.")
else:
    print("Verifikasi GAGAL: ada perbedaan signifikan.")
```

```python
# SOAL 11.5 - CHOLESKY DECOMPOSITION DAN PENYELESAIAN SISTEM

print("SOAL 11.5 - CHOLESKY DECOMPOSITION")
print()

A = [[6,   15,  55 ],
     [15,  55,  225],
     [55,  225, 979]]

r = [152.6, 585.6, 2488.8]

# LANGKAH 1: CHOLESKY DECOMPOSITION
print("Langkah 1: Cholesky Decomposition")
print()

l11 = (A[0][0]) ** 0.5
l21 = A[1][0] / l11
l31 = A[2][0] / l11
l22 = (A[1][1] - l21**2) ** 0.5
l32 = (A[2][1] - l31*l21) / l22
l33 = (A[2][2] - l31**2 - l32**2) ** 0.5

print(f"  l11 = sqrt({A[0][0]}) = {l11:.4f}")
print(f"  l21 = {A[1][0]} / {l11:.4f} = {l21:.4f}")
print(f"  l31 = {A[2][0]} / {l11:.4f} = {l31:.4f}")
print(f"  l22 = sqrt({A[1][1]} - {l21:.4f}^2) = {l22:.4f}")
print(f"  l32 = ({A[2][1]} - {l31:.4f}*{l21:.4f}) / {l22:.4f} = {l32:.4f}")
print(f"  l33 = sqrt({A[2][2]} - {l31:.4f}^2 - {l32:.4f}^2) = {l33:.4f}")
print()

print("Matriks L:")
print(f"  [{l11:.4f},  0,       0     ]")
print(f"  [{l21:.4f},  {l22:.4f},  0     ]")
print(f"  [{l31:.4f},  {l32:.4f},  {l33:.4f}]")
print()

# LANGKAH 2: FORWARD SUBSTITUTION [L]{d} = {r}
print("Langkah 2: Forward Substitution [L]{{d}} = {{r}}")
print()

d1 = r[0] / l11
d2 = (r[1] - l21*d1) / l22
d3 = (r[2] - l31*d1 - l32*d2) / l33

print(f"  d1 = {r[0]} / {l11:.4f} = {d1:.4f}")
print(f"  d2 = ({r[1]} - {l21:.4f}*{d1:.4f}) / {l22:.4f} = {d2:.4f}")
print(f"  d3 = ({r[2]} - {l31:.4f}*{d1:.4f} - {l32:.4f}*{d2:.4f}) / {l33:.4f} = {d3:.4f}")
print()

# LANGKAH 3: BACK SUBSTITUTION [L]^T{a} = {d}
print("Langkah 3: Back Substitution [L]^T{{a}} = {{d}}")
print()

a2 = d3 / l33
a1 = (d2 - l32*a2) / l22
a0 = (d1 - l21*a1 - l31*a2) / l11

print(f"  a2 = {d3:.4f} / {l33:.4f} = {a2:.4f}")
print(f"  a1 = ({d2:.4f} - {l32:.4f}*{a2:.4f}) / {l22:.4f} = {a1:.4f}")
print(f"  a0 = ({d1:.4f} - {l21:.4f}*{a1:.4f} - {l31:.4f}*{a2:.4f}) / {l11:.4f} = {a0:.4f}")
print()

print(f"Solusi: a0={a0:.4f}, a1={a1:.4f}, a2={a2:.4f}")
```

```python
# SOAL 11.6 - CHOLESKY DECOMPOSITION SISTEM 3x3

print("SOAL 11.6 - CHOLESKY DECOMPOSITION")
print()

A = [[8,  20, 15],
     [20, 80, 50],
     [15, 50, 60]]

r = [50, 250, 100]

# LANGKAH 1: CHOLESKY DECOMPOSITION
print("Langkah 1: Cholesky Decomposition")
print()

l11 = (A[0][0]) ** 0.5
l21 = A[1][0] / l11
l31 = A[2][0] / l11
l22 = (A[1][1] - l21**2) ** 0.5
l32 = (A[2][1] - l31*l21) / l22
l33 = (A[2][2] - l31**2 - l32**2) ** 0.5

print(f"  l11 = sqrt({A[0][0]}) = {l11:.4f}")
print(f"  l21 = {A[1][0]} / {l11:.4f} = {l21:.4f}")
print(f"  l31 = {A[2][0]} / {l11:.4f} = {l31:.4f}")
print(f"  l22 = sqrt({A[1][1]} - {l21:.4f}^2) = {l22:.4f}")
print(f"  l32 = ({A[2][1]} - {l31:.4f}*{l21:.4f}) / {l22:.4f} = {l32:.4f}")
print(f"  l33 = sqrt({A[2][2]} - {l31:.4f}^2 - {l32:.4f}^2) = {l33:.4f}")
print()

print("Matriks L:")
print(f"  [{l11:.4f},  0,       0     ]")
print(f"  [{l21:.4f},  {l22:.4f},  0     ]")
print(f"  [{l31:.4f},  {l32:.4f},  {l33:.4f}]")
print()

# LANGKAH 2: FORWARD SUBSTITUTION
print("Langkah 2: Forward Substitution [L]{{d}} = {{r}}")
print()

d1 = r[0] / l11
d2 = (r[1] - l21*d1) / l22
d3 = (r[2] - l31*d1 - l32*d2) / l33

print(f"  d1 = {r[0]} / {l11:.4f} = {d1:.4f}")
print(f"  d2 = ({r[1]} - {l21:.4f}*{d1:.4f}) / {l22:.4f} = {d2:.4f}")
print(f"  d3 = ({r[2]} - {l31:.4f}*{d1:.4f} - {l32:.4f}*{d2:.4f}) / {l33:.4f} = {d3:.4f}")
print()

# LANGKAH 3: BACK SUBSTITUTION
print("Langkah 3: Back Substitution [L]^T{{x}} = {{d}}")
print()

x3 = d3 / l33
x2 = (d2 - l32*x3) / l22
x1 = (d1 - l21*x2 - l31*x3) / l11

print(f"  x3 = {d3:.4f} / {l33:.4f} = {x3:.4f}")
print(f"  x2 = ({d2:.4f} - {l32:.4f}*{x3:.4f}) / {l22:.4f} = {x2:.4f}")
print(f"  x1 = ({d1:.4f} - {l21:.4f}*{x2:.4f} - {l31:.4f}*{x3:.4f}) / {l11:.4f} = {x1:.4f}")
print()

print(f"Solusi: x1={x1:.4f}, x2={x2:.4f}, x3={x3:.4f}")
```

```python
# SOAL 11.7 - CHOLESKY DECOMPOSITION MATRIKS DIAGONAL

print("SOAL 11.7 - CHOLESKY DECOMPOSITION MATRIKS DIAGONAL")
print()

A = [[9,  0,  0],
     [0, 25,  0],
     [0,  0,  4]]

# LANGKAH 1: CHOLESKY DECOMPOSITION
print("Langkah 1: Cholesky Decomposition")
print()

l11 = (A[0][0]) ** 0.5
l21 = A[1][0] / l11
l31 = A[2][0] / l11
l22 = (A[1][1] - l21**2) ** 0.5
l32 = (A[2][1] - l31*l21) / l22
l33 = (A[2][2] - l31**2 - l32**2) ** 0.5

print(f"  l11 = sqrt({A[0][0]}) = {l11:.4f}")
print(f"  l21 = {A[1][0]} / {l11:.4f} = {l21:.4f}")
print(f"  l31 = {A[2][0]} / {l11:.4f} = {l31:.4f}")
print(f"  l22 = sqrt({A[1][1]} - {l21:.4f}^2) = {l22:.4f}")
print(f"  l32 = ({A[2][1]} - {l31:.4f}*{l21:.4f}) / {l22:.4f} = {l32:.4f}")
print(f"  l33 = sqrt({A[2][2]} - {l31:.4f}^2 - {l32:.4f}^2) = {l33:.4f}")
print()

print("Matriks L:")
print(f"  [{l11:.4f},  0,      0     ]")
print(f"  [{l21:.4f},   {l22:.4f},  0     ]")
print(f"  [{l31:.4f},   {l32:.4f},  {l33:.4f}]")
print()

# LANGKAH 2: VERIFIKASI L x LT = A
print("Langkah 2: Verifikasi [L] x [L]^T = [A]")
print()

hasil = [[0.0]*3 for _ in range(3)]
LT = [[l11, l21, l31],
      [0,   l22, l32],
      [0,   0,   l33]]

L  = [[l11, 0,   0  ],
      [l21, l22, 0  ],
      [l31, l32, l33]]

for i in range(3):
    for j in range(3):
        hasil[i][j] = round(L[i][0]*LT[0][j] + L[i][1]*LT[1][j] + L[i][2]*LT[2][j], 4)

for row in hasil:
    print(f"  {row}")
print()

print("Matriks A asli:")
for row in A:
    print(f"  {row}")
print()
print("Hasil [L][L]^T sama dengan [A], dekomposisi terbukti benar.")
print()
print("Catatan: Karena A adalah matriks diagonal, maka L juga diagonal")
print(f"  l11=sqrt(9)={l11:.0f}, l22=sqrt(25)={l22:.0f}, l33=sqrt(4)={l33:.0f}")
```

```python
# SOAL 11.8 - GAUSS-SEIDEL DENGAN OVERRELAXATION

print("SOAL 11.8 - GAUSS-SEIDEL DENGAN OVERRELAXATION (λ=1.2)")
print()
print("Persamaan yang disederhanakan:")
print("  x1 = (41 + 0.4*x2) / 0.8")
print("  x2 = (25 + 0.4*x1 + 0.4*x3) / 0.8")
print("  x3 = (105 + 0.4*x2) / 0.8")
print()

lam = 1.2   # faktor overrelaxation
tol = 5     # toleransi error 5%

x1, x2, x3 = 0.0, 0.0, 0.0

for i in range(1, 100):
    x1_lama = x1
    x2_lama = x2
    x3_lama = x3

    # Hitung nilai baru
    x1_baru = (41 + 0.4*x2) / 0.8
    x2_baru = (25 + 0.4*x1_baru + 0.4*x3) / 0.8
    x3_baru = (105 + 0.4*x2_baru) / 0.8

    # Terapkan overrelaxation
    x1 = lam*x1_baru + (1-lam)*x1_lama
    x2 = lam*x2_baru + (1-lam)*x2_lama
    x3 = lam*x3_baru + (1-lam)*x3_lama

    ea1 = abs((x1 - x1_lama) / x1) * 100
    ea2 = abs((x2 - x2_lama) / x2) * 100
    ea3 = abs((x3 - x3_lama) / x3) * 100

    print(f"Iterasi {i}:")
    print(f"  x1 = {x1:.4f},  ea1 = {ea1:.4f}%")
    print(f"  x2 = {x2:.4f},  ea2 = {ea2:.4f}%")
    print(f"  x3 = {x3:.4f},  ea3 = {ea3:.4f}%")
    print()

    if max(ea1, ea2, ea3) < tol:
        print(f"Konvergen pada iterasi ke-{i}")
        break

print()
print(f"Solusi: x1={x1:.4f}, x2={x2:.4f}, x3={x3:.4f}")
```

```python
# SOAL 11.16a - SOLUSI, INVERS, DAN CONDITION NUMBER MATRIKS 3x3

print("SOAL 11.16a - MATRIKS 3x3")
print()

# LANGKAH 1: SOLUSI
print("Langkah 1: Solusi sistem persamaan")
print()

A = [[1,  4,  9],
     [4,  9, 16],
     [9, 16, 25]]

b = [14, 29, 50]

# Salin untuk dipakai ulang
A1 = [row[:] for row in A]
b1 = b[:]

faktor1 = A1[1][0] / A1[0][0]
faktor2 = A1[2][0] / A1[0][0]

A1[1][1] = A1[1][1] - faktor1 * A1[0][1]
A1[1][2] = A1[1][2] - faktor1 * A1[0][2]
b1[1]    = b1[1]    - faktor1 * b1[0]

A1[2][1] = A1[2][1] - faktor2 * A1[0][1]
A1[2][2] = A1[2][2] - faktor2 * A1[0][2]
b1[2]    = b1[2]    - faktor2 * b1[0]

faktor3  = A1[2][1] / A1[1][1]
A1[2][2] = A1[2][2] - faktor3 * A1[1][2]
b1[2]    = b1[2]    - faktor3 * b1[1]

x3 = b1[2] / A1[2][2]
x2 = (b1[1] - A1[1][2]*x3) / A1[1][1]
x1 = (b1[0] - A1[0][1]*x2 - A1[0][2]*x3) / A1[0][0]

print(f"  x1 = {x1:.4f}")
print(f"  x2 = {x2:.4f}")
print(f"  x3 = {x3:.4f}")
print()

# LANGKAH 2: INVERS
print("Langkah 2: Invers matriks menggunakan unit vector")
print()

def selesaikan(A, b):
    M = [row[:] for row in A]
    r = b[:]
    f1 = M[1][0] / M[0][0]
    f2 = M[2][0] / M[0][0]
    M[1][1] = M[1][1] - f1 * M[0][1]
    M[1][2] = M[1][2] - f1 * M[0][2]
    r[1]    = r[1]    - f1 * r[0]
    M[2][1] = M[2][1] - f2 * M[0][1]
    M[2][2] = M[2][2] - f2 * M[0][2]
    r[2]    = r[2]    - f2 * r[0]
    f3      = M[2][1] / M[1][1]
    M[2][2] = M[2][2] - f3 * M[1][2]
    r[2]    = r[2]    - f3 * r[1]
    z3 = r[2] / M[2][2]
    z2 = (r[1] - M[1][2]*z3) / M[1][1]
    z1 = (r[0] - M[0][1]*z2 - M[0][2]*z3) / M[0][0]
    return z1, z2, z3

k1 = selesaikan(A, [1, 0, 0])
k2 = selesaikan(A, [0, 1, 0])
k3 = selesaikan(A, [0, 0, 1])

inv = [[k1[0], k2[0], k3[0]],
       [k1[1], k2[1], k3[1]],
       [k1[2], k2[2], k3[2]]]

print("  [A]^-1 =")
for row in inv:
    print(f"    {[round(v, 4) for v in row]}")
print()

# LANGKAH 3: CONDITION NUMBER
print("Langkah 3: Condition number (row-sum norm)")
print()

norm_A   = max(sum(abs(A[i][j])   for j in range(3)) for i in range(3))
norm_inv = max(sum(abs(inv[i][j]) for j in range(3)) for i in range(3))
cond     = norm_A * norm_inv

print(f"  Norm [A]     = {norm_A:.4f}")
print(f"  Norm [A]^-1  = {norm_inv:.4f}")
print(f"  Condition number = {cond:.4f}")
```

```python
# SOAL 11.16b - SOLUSI, INVERS, DAN CONDITION NUMBER MATRIKS 4x4

print("SOAL 11.16b - MATRIKS 4x4")
print()

A = [[1,  4,  9, 16],
     [4,  9, 16, 25],
     [9, 16, 25, 36],
     [16, 25, 36, 49]]

b = [30, 54, 86, 126]

# LANGKAH 1: SOLUSI
print("Langkah 1: Solusi sistem persamaan")
print()

def gauss4(A, b):
    M = [row[:] for row in A]
    r = b[:]
    n = 4
    for k in range(n):
        for i in range(k+1, n):
            f = M[i][k] / M[k][k]
            for j in range(k, n):
                M[i][j] -= f * M[k][j]
            r[i] -= f * r[k]
    x = [0.0] * n
    for i in range(n-1, -1, -1):
        x[i] = (r[i] - sum(M[i][j]*x[j] for j in range(i+1, n))) / M[i][i]
    return x

sol = gauss4(A, b)
print(f"  x1 = {sol[0]:.4f}")
print(f"  x2 = {sol[1]:.4f}")
print(f"  x3 = {sol[2]:.4f}")
print(f"  x4 = {sol[3]:.4f}")
print()

# LANGKAH 2: INVERS
print("Langkah 2: Invers matriks menggunakan unit vector")
print()

k1 = gauss4(A, [1, 0, 0, 0])
k2 = gauss4(A, [0, 1, 0, 0])
k3 = gauss4(A, [0, 0, 1, 0])
k4 = gauss4(A, [0, 0, 0, 1])

inv = [[k1[0], k2[0], k3[0], k4[0]],
       [k1[1], k2[1], k3[1], k4[1]],
       [k1[2], k2[2], k3[2], k4[2]],
       [k1[3], k2[3], k3[3], k4[3]]]

print("  [A]^-1 =")
for row in inv:
    print(f"    {[round(v, 4) for v in row]}")
print()

# LANGKAH 3: CONDITION NUMBER
print("Langkah 3: Condition number (row-sum norm)")
print()

norm_A   = max(sum(abs(A[i][j])   for j in range(4)) for i in range(4))
norm_inv = max(sum(abs(inv[i][j]) for j in range(4)) for i in range(4))
cond     = norm_A * norm_inv

print(f"  Norm [A]     = {norm_A:.4f}")
print(f"  Norm [A]^-1  = {norm_inv:.4f}")
print(f"  Condition number = {cond:.4f}")
```

```python
# SOAL 11.17 - SISTEM PERSAMAAN NONLINEAR

print("SOAL 11.17 - SISTEM PERSAMAAN NONLINEAR")
print()
print("f(x, y) = 4 - y - 2*x^2 = 0")
print("g(x, y) = 8 - y^2 - 4*x = 0")
print()

def f(x, y):
    return 4 - y - 2*x**2

def g(x, y):
    return 8 - y**2 - 4*x

# BAGIAN (a) - CARI SOLUSI
print("Bagian (a) - Cari dua pasang solusi")
print()

def newton(x, y):
    for i in range(1, 100):
        fx = f(x, y)
        gx = g(x, y)

        dfx = -4*x
        dfy = -1
        dgx = -4
        dgy = -2*y

        det = dfx*dgy - dfy*dgx
        if abs(det) < 1e-10:
            break

        dx = (fx*dgy - gx*dfy) / det
        dy = (dfx*gx - dgx*fx) / det

        x = x - dx
        y = y - dy

        ea = max(abs(dx/(x+1e-10)), abs(dy/(y+1e-10))) * 100
        if ea < 0.001:
            break
    return round(x, 4), round(y, 4)

solusi1 = newton(1, 1)
solusi2 = newton(-1, 3)

print(f"  Solusi 1: x = {solusi1[0]}, y = {solusi1[1]}")
print(f"  Solusi 2: x = {solusi2[0]}, y = {solusi2[1]}")
print()

# BAGIAN (b) - BERBAGAI TEBAKAN AWAL
print("Bagian (b) - Tebakan awal x dan y dari -6 sampai 6")
print()
print(f"  {'x0':<6} {'y0':<6} {'x solusi':<12} {'y solusi':<12}")

for x0 in range(-6, 7, 2):
    for y0 in range(-6, 7, 2):
        try:
            xs, ys = newton(x0, y0)
            if abs(f(xs, ys)) < 0.01 and abs(g(xs, ys)) < 0.01:
                print(f"  {x0:<6} {y0:<6} {xs:<12} {ys:<12}")
        except:
            pass
```

```python
# SOAL 11.18 - SISTEM PRODUKSI ELEKTRONIK

print("SOAL 11.18 - SISTEM PRODUKSI ELEKTRONIK")
print()
print("Tabel kebutuhan bahan per komponen:")
print("  Transistor : Copper=4, Zinc=1, Glass=2")
print("  Resistor   : Copper=3, Zinc=3, Glass=1")
print("  Chip       : Copper=2, Zinc=1, Glass=3")
print()
print("Total bahan tersedia minggu ini:")
print("  Copper = 960")
print("  Zinc   = 510")
print("  Glass  = 610")
print()
print("Sistem persamaan:")
print("  4*T + 3*R + 2*C = 960")
print("  1*T + 3*R + 1*C = 510")
print("  2*T + 1*R + 3*C = 610")
print()

# Koefisien
a = [[4, 3, 2],
     [1, 3, 1],
     [2, 1, 3]]

b = [960, 510, 610]

# LANGKAH 1: ELIMINASI GAUSS
print("Langkah 1: Eliminasi Gauss")
print()

# Eliminasi kolom 1
faktor1 = a[1][0] / a[0][0]
faktor2 = a[2][0] / a[0][0]

a[1][1] = a[1][1] - faktor1 * a[0][1]
a[1][2] = a[1][2] - faktor1 * a[0][2]
b[1]    = b[1]    - faktor1 * b[0]

a[2][1] = a[2][1] - faktor2 * a[0][1]
a[2][2] = a[2][2] - faktor2 * a[0][2]
b[2]    = b[2]    - faktor2 * b[0]

print(f"  faktor1 = {faktor1:.4f}")
print(f"  faktor2 = {faktor2:.4f}")
print(f"  Baris 2 baru: {[round(v,4) for v in a[1]]} = {round(b[1],4)}")
print(f"  Baris 3 baru: {[round(v,4) for v in a[2]]} = {round(b[2],4)}")
print()

# Eliminasi kolom 2
faktor3 = a[2][1] / a[1][1]

a[2][2] = a[2][2] - faktor3 * a[1][2]
b[2]    = b[2]    - faktor3 * b[1]

print(f"  faktor3 = {faktor3:.4f}")
print(f"  Baris 3 baru: {[round(v,4) for v in a[2]]} = {round(b[2],4)}")
print()

# LANGKAH 2: BACK SUBSTITUTION
print("Langkah 2: Back Substitution")
print()

C = b[2] / a[2][2]
R = (b[1] - a[1][2]*C) / a[1][1]
T = (b[0] - a[0][1]*R - a[0][2]*C) / a[0][0]

print(f"  C = {b[2]:.4f} / {a[2][2]:.4f} = {C:.4f}")
print(f"  R = ({b[1]:.4f} - {a[1][2]:.4f}*{C:.4f}) / {a[1][1]:.4f} = {R:.4f}")
print(f"  T = ({b[0]:.4f} - {a[0][1]:.4f}*{R:.4f} - {a[0][2]:.4f}*{C:.4f}) / {a[0][0]:.4f} = {T:.4f}")
print()

print(f"Solusi:")
print(f"  Transistor (T) = {round(T)} unit")
print(f"  Resistor   (R) = {round(R)} unit")
print(f"  Chip       (C) = {round(C)} unit")
```

```python
# SOAL 11.19 - HILBERT MATRIX 10x10

print("SOAL 11.19 - HILBERT MATRIX 10x10")
print()

n = 10

# LANGKAH 1: BUAT MATRIKS HILBERT
print("Langkah 1: Buat Matriks Hilbert 10x10")
print()

H = [[1/(i+j+1) for j in range(n)] for i in range(n)]

print("  Matriks Hilbert H (5 baris pertama):")
for i in range(5):
    print(f"    {[round(H[i][j], 4) for j in range(n)]}")
print("  ...")
print()

# LANGKAH 2: RUAS KANAN (jumlah tiap baris)
print("Langkah 2: Ruas kanan = jumlah tiap baris")
print()

b = [sum(H[i][j] for j in range(n)) for i in range(n)]
print("  b =", [round(v, 6) for v in b])
print()
print("  Catatan: solusi sebenarnya semua x = 1")
print()

# FUNGSI GAUSS UMUM
def gauss(A, b):
    M = [row[:] for row in A]
    r = b[:]
    n = len(r)
    for k in range(n):
        for i in range(k+1, n):
            if M[k][k] == 0:
                continue
            f = M[i][k] / M[k][k]
            for j in range(k, n):
                M[i][j] -= f * M[k][j]
            r[i] -= f * r[k]
    x = [0.0] * n
    for i in range(n-1, -1, -1):
        x[i] = (r[i] - sum(M[i][j]*x[j] for j in range(i+1, n))) / M[i][i]
    return x

# LANGKAH 3: SOLUSI
print("Langkah 3: Solusi sistem")
print()

sol = gauss(H, b)
print("  Solusi yang dihitung:")
for i in range(n):
    print(f"    x{i+1} = {sol[i]:.6f}")
print()

# LANGKAH 4: INVERS
print("Langkah 4: Hitung invers menggunakan unit vector")
print()

inv = [[0.0]*n for _ in range(n)]
for j in range(n):
    e = [1.0 if i == j else 0.0 for i in range(n)]
    col = gauss(H, e)
    for i in range(n):
        inv[i][j] = col[i]

print("  Invers H (5 baris, 5 kolom pertama):")
for i in range(5):
    print(f"    {[round(inv[i][j], 2) for j in range(5)]}")
print("  ...")
print()

# LANGKAH 5: CONDITION NUMBER
print("Langkah 5: Condition number (row-sum norm)")
print()

norm_H   = max(sum(abs(H[i][j])   for j in range(n)) for i in range(n))
norm_inv = max(sum(abs(inv[i][j]) for j in range(n)) for i in range(n))
cond     = norm_H * norm_inv

print(f"  Norm [H]     = {norm_H:.6f}")
print(f"  Norm [H]^-1  = {norm_inv:.2f}")
print(f"  Condition number = {cond:.2f}")
print()

# LANGKAH 6: DIGIT PRESISI YANG HILANG
print("Langkah 6: Digit presisi yang hilang")
print()

import math
digit_hilang = math.log10(cond)
print(f"  log10({cond:.2f}) = {digit_hilang:.2f}")
print(f"  Sekitar {digit_hilang:.0f} digit presisi hilang akibat ill-conditioning")
print()

# LANGKAH 7: ERROR SOLUSI
print("Langkah 7: Bandingkan solusi dengan nilai sebenarnya (semua x=1)")
print()

print(f"  {'x':<6} {'Hitung':<15} {'Sebenarnya':<12} {'Error':<10}")
for i in range(n):
    error = abs(sol[i] - 1.0)
    print(f"  x{i+1:<5} {sol[i]:<15.6f} {1.0:<12} {error:<10.6f}")
```

```python
# SOAL 11.20 - VANDERMONDE MATRIX 6x6

print("SOAL 11.20 - VANDERMONDE MATRIX 6x6")
print()

n = 6
x = [1, 2, 3, 4, 5, 6]

# LANGKAH 1: BUAT MATRIKS VANDERMONDE
print("Langkah 1: Buat Matriks Vandermonde 6x6")
print()
print(f"  x = {x}")
print()

V = [[x[i]**(n-1-j) for j in range(n)] for i in range(n)]

print("  Matriks Vandermonde V:")
for row in V:
    print(f"    {row}")
print()

# LANGKAH 2: RUAS KANAN (jumlah tiap baris)
print("Langkah 2: Ruas kanan = jumlah tiap baris")
print()

b = [sum(V[i][j] for j in range(n)) for i in range(n)]
print("  b =", b)
print()
print("  Catatan: solusi sebenarnya semua x = 1")
print()

# FUNGSI GAUSS UMUM
def gauss(A, b):
    M = [row[:] for row in A]
    r = b[:]
    n = len(r)
    for k in range(n):
        for i in range(k+1, n):
            if M[k][k] == 0:
                continue
            f = M[i][k] / M[k][k]
            for j in range(k, n):
                M[i][j] -= f * M[k][j]
            r[i] -= f * r[k]
    x = [0.0] * n
    for i in range(n-1, -1, -1):
        x[i] = (r[i] - sum(M[i][j]*x[j] for j in range(i+1, n))) / M[i][i]
    return x

# LANGKAH 3: SOLUSI
print("Langkah 3: Solusi sistem")
print()

sol = gauss(V, b)
print("  Solusi yang dihitung:")
for i in range(n):
    print(f"    x{i+1} = {sol[i]:.6f}")
print()

# LANGKAH 4: INVERS
print("Langkah 4: Hitung invers menggunakan unit vector")
print()

inv = [[0.0]*n for _ in range(n)]
for j in range(n):
    e = [1.0 if i == j else 0.0 for i in range(n)]
    col = gauss(V, e)
    for i in range(n):
        inv[i][j] = col[i]

print("  Invers V:")
for row in inv:
    print(f"    {[round(v, 4) for v in row]}")
print()

# LANGKAH 5: CONDITION NUMBER
print("Langkah 5: Condition number (row-sum norm)")
print()

norm_V   = max(sum(abs(V[i][j])   for j in range(n)) for i in range(n))
norm_inv = max(sum(abs(inv[i][j]) for j in range(n)) for i in range(n))
cond     = norm_V * norm_inv

print(f"  Norm [V]     = {norm_V:.4f}")
print(f"  Norm [V]^-1  = {norm_inv:.4f}")
print(f"  Condition number = {cond:.4f}")
print()

# LANGKAH 6: DIGIT PRESISI YANG HILANG
print("Langkah 6: Digit presisi yang hilang")
print()

import math
digit_hilang = math.log10(cond)
print(f"  log10({cond:.2f}) = {digit_hilang:.2f}")
print(f"  Sekitar {digit_hilang:.0f} digit presisi hilang akibat ill-conditioning")
print()

# LANGKAH 7: ERROR SOLUSI
print("Langkah 7: Bandingkan solusi dengan nilai sebenarnya (semua x=1)")
print()

print(f"  {'x':<6} {'Hitung':<15} {'Sebenarnya':<12} {'Error':<10}")
for i in range(n):
    error = abs(sol[i] - 1.0)
    print(f"  x{i+1:<5} {sol[i]:<15.6f} {1.0:<12} {error:<10.6f}")
```

```python
# SOAL 11.22 - SOLUSI, TRANSPOSE, DAN INVERS

print("SOAL 11.22 - SOLUSI, TRANSPOSE, DAN INVERS")
print()
print("Sistem persamaan:")
print("  50  = 5*x1 - 7*x2")
print("  4*x2 + 7*x3 + 30 = 0  ->  4*x2 + 7*x3 = -30")
print("  x1 - 7*x3 = 40 - 3*x2 + 5*x1  ->  -4*x1 + 3*x2 - 7*x3 = 40")
print()

# Matriks koefisien
A = [[ 5, -7,  0],
     [ 0,  4,  7],
     [-4,  3, -7]]

b = [50, -30, 40]

print("Matriks A:")
for row in A:
    print(f"  {row}")
print()

# LANGKAH 1: ELIMINASI GAUSS
print("Langkah 1: Eliminasi Gauss")
print()

faktor1 = A[1][0] / A[0][0]
faktor2 = A[2][0] / A[0][0]

A[1][1] = A[1][1] - faktor1 * A[0][1]
A[1][2] = A[1][2] - faktor1 * A[0][2]
b[1]    = b[1]    - faktor1 * b[0]

A[2][1] = A[2][1] - faktor2 * A[0][1]
A[2][2] = A[2][2] - faktor2 * A[0][2]
b[2]    = b[2]    - faktor2 * b[0]

faktor3 = A[2][1] / A[1][1]
A[2][2] = A[2][2] - faktor3 * A[1][2]
b[2]    = b[2]    - faktor3 * b[1]

print(f"  faktor1 = {faktor1:.4f}")
print(f"  faktor2 = {faktor2:.4f}")
print(f"  faktor3 = {faktor3:.4f}")
print()

# LANGKAH 2: BACK SUBSTITUTION
print("Langkah 2: Back Substitution")
print()

x3 = b[2] / A[2][2]
x2 = (b[1] - A[1][2]*x3) / A[1][1]
x1 = (b[0] - A[0][1]*x2 - A[0][2]*x3) / A[0][0]

print(f"  x3 = {x3:.4f}")
print(f"  x2 = {x2:.4f}")
print(f"  x1 = {x1:.4f}")
print()

# LANGKAH 3: TRANSPOSE
print("Langkah 3: Transpose matriks A")
print()

A_orig = [[ 5, -7,  0],
          [ 0,  4,  7],
          [-4,  3, -7]]

AT = [[A_orig[j][i] for j in range(3)] for i in range(3)]

print("  Matriks A^T:")
for row in AT:
    print(f"    {row}")
print()

# LANGKAH 4: INVERS
print("Langkah 4: Invers matriks A menggunakan unit vector")
print()

def gauss_solve(M, b):
    A = [row[:] for row in M]
    b = b[:]
    n = len(b)
    for k in range(n):
        for i in range(k+1, n):
            f = A[i][k] / A[k][k]
            for j in range(k, n):
                A[i][j] -= f * A[k][j]
            b[i] -= f * b[k]
    x = [0.0] * n
    for i in range(n-1, -1, -1):
        x[i] = (b[i] - sum(A[i][j]*x[j] for j in range(i+1, n))) / A[i][i]
    return x

A_orig = [[ 5, -7,  0],
          [ 0,  4,  7],
          [-4,  3, -7]]

e1 = gauss_solve(A_orig, [1, 0, 0])
e2 = gauss_solve(A_orig, [0, 1, 0])
e3 = gauss_solve(A_orig, [0, 0, 1])

print("  Matriks Invers [A]^-1:")
print(f"    {[round(v,4) for v in e1]}")
print(f"    {[round(v,4) for v in e2]}")
print(f"    {[round(v,4) for v in e3]}")
```

```python
# SOAL 11.24 - PROGRAM THOMAS ALGORITHM (duplikasi Example 11.1)

print("SOAL 11.24 - THOMAS ALGORITHM")
print()
print("Duplikasi Example 11.1")
print()

e = [0,    -1,    -1   ]
f = [2.04,  2.04,  2.04]
g = [-1,   -1,     0   ]
r = [40.8,  0.8,   200.8]

n = 3

print("Langkah 1: Forward Decomposition")
print()

for k in range(1, n):
    faktor = e[k] / f[k-1]
    f[k] = f[k] - faktor * g[k-1]
    r[k] = r[k] - faktor * r[k-1]
    print(f"  k={k+1}: faktor={faktor:.4f}, f[{k}]={f[k]:.4f}, r[{k}]={r[k]:.4f}")

print()
print("Langkah 2: Back Substitution")
print()

T3 = r[2] / f[2]
T2 = (r[1] - g[1]*T3) / f[1]
T1 = (r[0] - g[0]*T2) / f[0]

print(f"  T3 = {r[2]:.4f} / {f[2]:.4f} = {T3:.4f}")
print(f"  T2 = ({r[1]:.4f} - ({g[1]})*{T3:.4f}) / {f[1]:.4f} = {T2:.4f}")
print(f"  T1 = ({r[0]:.4f} - ({g[0]})*{T2:.4f}) / {f[0]:.4f} = {T1:.4f}")
print()
print(f"Solusi: T1={T1:.4f}, T2={T2:.4f}, T3={T3:.4f}")
```

```python
# SOAL 11.25 - PROGRAM CHOLESKY DECOMPOSITION (duplikasi Example 11.2)

print("SOAL 11.25 - CHOLESKY DECOMPOSITION")
print()
print("Duplikasi Example 11.2")
print()

A = [[6,   15,  55 ],
     [15,  55,  225],
     [55,  225, 979]]

print("Matriks A:")
for row in A:
    print(f"  {row}")
print()

print("Langkah 1: Cholesky Decomposition")
print()

l11 = A[0][0] ** 0.5
l21 = A[1][0] / l11
l31 = A[2][0] / l11
l22 = (A[1][1] - l21**2) ** 0.5
l32 = (A[2][1] - l31*l21) / l22
l33 = (A[2][2] - l31**2 - l32**2) ** 0.5

print(f"  l11 = sqrt({A[0][0]}) = {l11:.4f}")
print(f"  l21 = {A[1][0]} / {l11:.4f} = {l21:.4f}")
print(f"  l31 = {A[2][0]} / {l11:.4f} = {l31:.4f}")
print(f"  l22 = sqrt({A[1][1]} - {l21:.4f}^2) = {l22:.4f}")
print(f"  l32 = ({A[2][1]} - {l31:.4f}*{l21:.4f}) / {l22:.4f} = {l32:.4f}")
print(f"  l33 = sqrt({A[2][2]} - {l31:.4f}^2 - {l32:.4f}^2) = {l33:.4f}")
print()

print("Matriks L:")
print(f"  [{l11:.4f},  0,       0     ]")
print(f"  [{l21:.4f},  {l22:.4f},  0     ]")
print(f"  [{l31:.4f},  {l32:.4f},  {l33:.4f}]")
```

```python
# SOAL 11.26 - PROGRAM GAUSS-SEIDEL (duplikasi Example 11.3)

print("SOAL 11.26 - GAUSS-SEIDEL")
print()
print("Duplikasi Example 11.3")
print()
print("Persamaan yang disederhanakan:")
print("  x1 = (7.85 + 0.1*x2 + 0.2*x3) / 2.04")
print("  x2 = (-19.3 + 0.1*x1 + 0.3*x3) / 2.04")
print("  x3 = (71.4 - 0.1*x1 + 0.2*x2) / 1.5")
print()

x1, x2, x3 = 0.0, 0.0, 0.0
tol = 5

for i in range(1, 100):
    x1_lama = x1
    x2_lama = x2
    x3_lama = x3

    x1 = (7.85  + 0.1*x2  + 0.2*x3) / 2.04
    x2 = (-19.3 + 0.1*x1  + 0.3*x3) / 2.04
    x3 = (71.4  - 0.1*x1  + 0.2*x2) / 1.5

    ea1 = abs((x1 - x1_lama) / x1) * 100 if x1 != 0 else float('inf')
    ea2 = abs((x2 - x2_lama) / x2) * 100 if x2 != 0 else float('inf')
    ea3 = abs((x3 - x3_lama) / x3) * 100 if x3 != 0 else float('inf')

    print(f"Iterasi {i}:")
    print(f"  x1 = {x1:.4f},  ea1 = {ea1:.4f}%")
    print(f"  x2 = {x2:.4f},  ea2 = {ea2:.4f}%")
    print(f"  x3 = {x3:.4f},  ea3 = {ea3:.4f}%")
    print()

    if max(ea1, ea2, ea3) < tol:
        print(f"Konvergen pada iterasi ke-{i}")
        break

print()
print(f"Solusi: x1={x1:.4f}, x2={x2:.4f}, x3={x3:.4f}")
```

```python
# SOAL 11.27 - PERSAMAAN DIFERENSIAL STEADY-STATE

print("SOAL 11.27 - PERSAMAAN DIFERENSIAL STEADY-STATE")
print()
print("Persamaan diferensial:")
print("  D*d2c/dx2 - U*dc/dx - k*c = 0")
print()
print("Parameter:")
print("  D=2, U=1, k=0.2, c(0)=80, c(10)=20, dx=2")
print()
print("Diskretisasi dengan beda hingga:")
print("  D*(c[i+1]-2*c[i]+c[i-1])/dx^2 - U*(c[i+1]-c[i-1])/(2*dx) - k*c[i] = 0")
print()

D  = 2
U  = 1
k  = 0.2
dx = 2
c0 = 80
cL = 20

# Koefisien diskretisasi
a = D/dx**2 + U/(2*dx)   # koefisien c[i-1]
b = -2*D/dx**2 - k       # koefisien c[i]
c = D/dx**2 - U/(2*dx)   # koefisien c[i+1]

print(f"Koefisien:")
print(f"  a (c[i-1]) = {a:.4f}")
print(f"  b (c[i])   = {b:.4f}")
print(f"  c (c[i+1]) = {c:.4f}")
print()

# Node: x=0,2,4,6,8,10 -> 6 titik, 4 titik interior
# Sistem 4 persamaan untuk c1,c2,c3,c4
# c0=80 dan c5=20 diketahui

r = [-a*c0, 0, 0, -c*cL]

A = [[b, c, 0, 0],
     [a, b, c, 0],
     [0, a, b, c],
     [0, 0, a, b]]

print("Langkah 1: Eliminasi Gauss")
print()

n = 4
for k_idx in range(n):
    for i in range(k_idx+1, n):
        if A[k_idx][k_idx] == 0:
            continue
        f = A[i][k_idx] / A[k_idx][k_idx]
        for j in range(k_idx, n):
            A[i][j] -= f * A[k_idx][j]
        r[i] -= f * r[k_idx]
        print(f"  faktor({i+1},{k_idx+1}) = {f:.4f}")

print()
print("Langkah 2: Back Substitution")
print()

sol = [0.0] * n
for i in range(n-1, -1, -1):
    sol[i] = (r[i] - sum(A[i][j]*sol[j] for j in range(i+1, n))) / A[i][i]
    print(f"  c{i+1} = {sol[i]:.4f}")

print()
print("Konsentrasi vs Jarak:")
print(f"  {'x':<6} {'c':<10}")
print(f"  {0:<6} {c0:<10}")
for i in range(4):
    print(f"  {(i+1)*2:<6} {round(sol[i],4):<10}")
print(f"  {10:<6} {cL:<10}")
```

```python
# SOAL 11.28 - SISTEM PENTADIAGONAL

print("SOAL 11.28 - SISTEM PENTADIAGONAL")
print()

# Koefisien pentadiagonal
# f = diagonal utama
# g = superdiagonal 1, h = superdiagonal 2
# e = subdiagonal 1, d = subdiagonal 2

f = [ 8,  9,  7, 12, 15]  # diagonal utama
g = [-2, -4, -1, -5,  0]  # superdiagonal 1
h = [-1, -1, -2,  0,  0]  # superdiagonal 2
e = [ 0, -2, -3, -4, -7]  # subdiagonal 1
d = [ 0,  0, -1,  0,  0]  # subdiagonal 2
r = [ 5,  2,  0,  1,  5]  # ruas kanan

n = 5

print("Matriks pentadiagonal:")
print(f"  f (diagonal utama)  = {f}")
print(f"  g (superdiagonal 1) = {g}")
print(f"  h (superdiagonal 2) = {h}")
print(f"  e (subdiagonal 1)   = {e}")
print(f"  d (subdiagonal 2)   = {d}")
print(f"  r (ruas kanan)      = {r}")
print()

# Buat matriks penuh untuk eliminasi Gauss
A = [[0.0]*n for _ in range(n)]
for i in range(n):
    A[i][i] = f[i]
    if i < n-1:
        A[i][i+1] = g[i]
    if i < n-2:
        A[i][i+2] = h[i]
    if i > 0:
        A[i][i-1] = e[i]
    if i > 1:
        A[i][i-2] = d[i]

print("Langkah 1: Eliminasi Gauss")
print()

for k in range(n):
    for i in range(k+1, min(k+3, n)):
        faktor = A[i][k] / A[k][k]
        for j in range(k, n):
            A[i][j] -= faktor * A[k][j]
        r[i] -= faktor * r[k]
        print(f"  faktor({i+1},{k+1}) = {faktor:.4f}")

print()
print("Langkah 2: Back Substitution")
print()

x = [0.0] * n
for i in range(n-1, -1, -1):
    x[i] = (r[i] - sum(A[i][j]*x[j] for j in range(i+1, n))) / A[i][i]
    print(f"  x{i+1} = {x[i]:.4f}")

print()
print(f"Solusi: x1={x[0]:.4f}, x2={x[1]:.4f}, x3={x[2]:.4f}, x4={x[3]:.4f}, x5={x[4]:.4f}")
```
